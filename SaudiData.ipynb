```python
import pandas as pd
import random
# generate synthetic job data
industries = ["IT","Finance", "Construction", "Healtcare", "Retail"]
locations = ["Riyadh", "Jeddah", "Dammam", "Mecca", "Medina"]
skills = ["Python", "Excel", "Project Managment", "Data Analysis", "AI/ML"]
# Create 500 synthetic job postings
data = []
for _ in range(500):
  job = {
      "Title": random.choice (["Data analysis", "Software Engineer", "Accountant", "Project Manager"]),
      "Industry":random.choice(industries),
      "Location": random.choice(locations),
      "Skills": random.sample(skills, k=2),
      "Salary": random.randint(5000, 20000),   # Monthlly salary in SAR
      "Date_Posted": pd.Timestamp("2025-01-01") - pd.to_timedelta(random.randint(1, 365), unit='d')
  }
  data.append(job)
# Convert to DataFrame
df_synthetic_jobs = pd.DataFrame(data)

# Save synthetic data
df_synthetic_jobs.to_csv("synthetic_jobs_data_saudi,csv", index=False)
print("Synthetic job data generated and saved!")
```

```python
total_jobs = df_synthetic_jobs.shape[0]
avg_salary = df_synthetic_jobs["Salary"].mean()
date_min =pd.to_datetime(df_synthetic_jobs["Date_Posted"]).min()
date_max = pd.to_datetime(df_synthetic_jobs["Date_Posted"]).max()

print(f"Total Job Postings: {total_jobs}")
print(f"Average Salary: {avg_salary:.2f} SAR")
print(f"Date Range: {date_min.date()} to {date_max.date()}")
```

```python
industry_counts = df_synthetic_jobs ["Industry"].value_counts()
industry_avg_salary = df_synthetic_jobs.groupby("Industry")["Salary"].mean()
print("Job Postings by Industry:")
print("\nAverage Salary by Industry (SAR):")
print(industry_avg_salary)
```
